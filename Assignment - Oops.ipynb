```python
# Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance.
#           Implement methods to deposit and withdraw money from the account.


class Bank_account :
    
    def __init__(self, accountNumber, acoountHolderName, balance):
        self.account_number = accountNumber
        self.account_holder_name = acoountHolderName
        self.balance = balance
    
    def deposit(self):  # Create menthod deposit money to bank account
        
        amount = int(input("Enter the deposit amount : "))
        self.balance += amount
        print("Amount Deposited : ", amount)
        print("Available Balance is : ", self.balance)
        
    def withdraw(self):  # Create method to withdraw money from bank account
        
        amount = int(input("Enter the withdraw amount : "))
        if self.balance >= amount:
            self.balance -= amount
            print("Amount withdraw : ", amount)
            print("Available Balance is : ", self.balance)
        else:
            print("Insufficient Amount")

obj1 = Bank_account(12334, "farhan", 500)  # creating an object for class

obj1.deposit()   # calling function with that class object
obj1.withdraw()

```

```python
# Problem 3: Vehicle Rental Create a class representing a vehicle rental system.
#            Implement methods to rent a vehicle, return a vehicle, and display available vehicles.

class Vehicle:
    def __init__(self, vehicleID, model, available = True):
        self.vehicleID = vehicleID
        self.model = model
        self.available = available

class vehicle_rental :
    def __init__(self):
        self.vehicles = {}
    
    def add_vehicle(self, vehicleID, model):
        vehicle = Vehicle(vehicleID, model)
        self.vehicles[vehicleID] = vehicle

    def rent_vehicle(self, vehicleID):
        if vehicleID in self.vehicles:
            if self.vehicles[vehicleID].available:
                self.vehicles[vehicleID].available = False
                print(f"\n Vehicle {vehicleID} rented successfully.")
            else:
                print(f"\n Vehicle {vehicleID} is not available for rent.")
        else:
            print(f"\n Vehicle {vehicleID} not found.")
        
    def return_vehicle(self, vehicleID):
        if vehicleID in self.vehicles:
            if not self.vehicles[vehicleID].available:
                self.vehicles[vehicleID].available = True
                print(f"\n Vehicle {vehicleID} returned successfully.")
            else:
                print(f"\n Vehicle {vehicleID} is already available.")
        else:
            print(f"\n Vehicle {vehicleID} not found.")
            
    def display_available_vehicle(self):
        available_vehicles = [vehicle_id for vehicle_id, vehicle in self.vehicles.items() if vehicle.available]
        if available_vehicles:
            print(" Available Vehicles: ")
            for vehicle_id in available_vehicles:
                print(f"{vehicle_id} - {self.vehicles[vehicle_id].model}")
        else:
            print("No vehicles available.")
            
            
V1 = vehicle_rental()

# Adding vehicles to the system
V1.add_vehicle("R15", "Bajaj")
V1.add_vehicle("KTM", "Kawasaki")
V1.add_vehicle("Ninja", "Kawasaki")
V1.add_vehicle("CBR", "Honda")


V1.display_available_vehicle()

# Rent a vehicle
V1.rent_vehicle("R15")

# Display available vehicles after renting
V1.display_available_vehicle()

# Return a vehicle
V1.return_vehicle("R15")

# Display available vehicles after returning
V1.display_available_vehicle()

```

```python
# Problem 4: Library Catalog Create classes representing a library and a book.
#            Implement methods to add books to the library, borrow books, and display available books.

class Library:
    def __init__(self, book_id, book_name, available = True):
        self.book_id = book_id
        self.book_name = book_name
        self.available = available
        
class library_catalog:
    def __init__(self):
        self.books = {}
        
    def add_books(self, book_id, book_name):
        book = Library(book_id, book_name)
        self.books[book_id] = book
        
    def borrow_book(self, book_id):
        if book_id in self.books:
            if self.books[book_id].available:
                self.books[book_id].available = False
                print(f"\n BOOK {book_id} borrow successfully.")
            else:
                print(f"\n BOOK {book_id} is not available for borrow.")
        else:
            print(f"\n BOOK {book_id} not found.")
            
    def return_book(self, book_id):
        if book_id in self.books:
            if not self.books[book_id].available:
                self.books[book_id].available = True
                print(f"\n BOOK {book_id} returned successfully.")
            else:
                print(f"\n BOOK {book_id} is already available.")
        else:
            print(f"\n BOOK {book_id} not found.")
            
    def display_available_book(self):
        available_books = [book_id for book_id, book in self.books.items() if book.available]
        if available_books:
            print(" Available BOOK: ")
            for book_id in available_books:
                print(f"{book_id} - {self.books[book_id].book_name}")
        else:
            print("No BOOK available.")

B1 = library_catalog()
B1.add_books("Physics1", "HC VERMA")
B1.add_books("Math1", "RD SHARMA")
B1.add_books("Chemistry1", "BALAJI")
B1.add_books("Physics2", "UNIVERSAL")
B1.add_books("Math2", "MANOHAR RE")

B1.display_available_book()
B1.borrow_book("Math1")
B1.borrow_book("Math2")

B1.display_available_book()
  
B1.return_book("Math1")

B1.display_available_book()
```

```python
# Problem 5: Product Inventory Create classes representing a product and an inventory system.
#            Implement methods to add products to the inventory, update product quantity, and display available products.

class Product:
    def __init__(self, product_id, name, price, quantity):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity
    
    def update_quantity(self, new_quantity):
        if new_quantity >= 0:
            self.quantity = new_quantity
            print(f"\n Quantity for {self.name} updated to {self.quantity}.")
        else:
            print("\n Invalid quantity. Quantity must be a non-negative number.")
            
    def __str__(self):
        return f"{self.name} - Price: {self.price} - Quantity: {self.quantity}"
    
class inventory_system:
    def __init__(self):
        self.products = {}
        
    def add_product(self, product_id, name, price, quantity):
        if product_id not in self.products:
            product = Product(product_id, name, price, quantity)
            self.products[product_id] = product
            print(f"{name} added to the inventory.")
        else:
            print(f"Product with ID {product_id} already exists in the inventory.")
        
    def update_product_quantity(self, product_id, new_quantity):
        if product_id in self.products:
            self.products[product_id].update_quantity(new_quantity)
        else:
            print(f"Product with ID {product_id} not found in the inventory.")
            
    def display_available_product(self):
        if self.products:
            print("\n Available Products")
            for product in self.products.values():
                if product.quantity > 0:
                    print(product)
        else:
            print("No Product in the Inventory \n")
            
Product1 = inventory_system()           # create inventory_system object
Product1.display_available_product()    # display product

# Add products
Product1.add_product("p1", "Oppo A11", 12000, 5)
Product1.add_product("p2", "1 Plus Nord", 20000, 3)
Product1.add_product("p3", "Realme 12", 13000, 5)
Product1.add_product("p4", "Apple iphone 15", 90000, 2)
Product1.add_product("p5", "Apple iphone 15 pro", 120000, 2)

Product1.display_available_product()    # display product

Product1.update_product_quantity("p4", 4)

Product1.display_available_product()    # display product
```

```python
# Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height.
#            Implement methods to calculate the area and perimeter of the shape.

class Shape:
    def __init__(self, length, width, height):
        self.length = length
        self.width = width
        self.height = height
        
    def calculate_area_perimeter(self):
        if (self.length == self.width == self.height):
            print("Cube Shape")
            area = 6*self.length*self.length
            perimeter = 12*self.length
            print(f"Area is : {area} and Perimeter is : {perimeter} \n")
            
        else:
            print("Cuboid Shape")
            area = 2*(self.length*self.width + self.width*self.height + self.height*self.length)
            perimeter = 4*(self.length + self.width +self.height)
            print(f"Area is : {area} and Perimeter is : {perimeter} \n")
            
# create object with attribute length, width, hw=eight
S = Shape(2, 3, 4)
S.calculate_area_perimeter()    # calculate area and perimeter

S1 = Shape(2, 2, 4)
S1.calculate_area_perimeter()

S2 = Shape(2, 2, 2)
S2.calculate_area_perimeter()
```

```python
# Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades.
#            Implement methods to calculate the average grade and display student details.

class Student:
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name
        self.grades = []
        
    def add_grade(self, grade):
        if 0 <= grade <= 100:
            self.grades.append(grade)
            print(f"Grade {grade} added for {self.name}.")
            
    def calculate_average_grade(self):
        if self.grades:
            average = sum(self.grades)/len(self.grades)
            return average
        else:
            return 0
    
    def display_student_details(self):
        print(f"Student ID: {self.student_id}")
        print(f"Name: {self.name}")
        if self.grades:
            print("Grades:")
            for grade in self.grades:
                print(f"- {grade}")
            print(f"Average Grade: {self.calculate_average_grade()}")
        else:
            print("No grades available for this student.")
            
            
S1 = Student("GL3147", "Farhan")
S2 = Student("GL3144", "Riyozaki")

S1.add_grade(90)
S1.add_grade(86)
S1.add_grade(96)
S1.add_grade(76)
S2.add_grade(96)
S2.add_grade(94)
S2.add_grade(98)
S2.add_grade(99)

S1.display_student_details()
print()
S2.display_student_details()
```

```python
# Problem 8: Email Management Create a class representing an email with attributes like sender, recipient, and subject.
#            Implement methods to send an email and display email details.

class Email:
    def __init__(self, sender, recipient, subject):
        self.sender = sender
        self.recipient = recipient
        self.subject = subject
        self.sent = False
        
    def send_mail(self):
        if not self.sent:
            self.sent = True
            print(f"Email sent from {self.sender} to {self.recipient}")
        else:
            print("This email has already been sent")
            
    def display_email_details(self):
        print("Email Details:")
        print(f"From: {self.sender}")
        print(f"To: {self.recipient}")
        print(f"Subject: {self.subject}")

E1 = Email("fmohd@gmail.com", "ayush@gmail.com", "Bunk")
E2 = Email("riyo@gmail.com", "ayu@gmail.com", "Bunk Cancel")

E1.display_email_details()
E1.send_mail()
E1.send_mail()

print()
E2.display_email_details()
E2.send_mail()
```

```python
# Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts.
#            Implement methods to add posts, display posts, and search for posts by keyword.

class Social_media:
    def __init__(self, username):
        self.username = username
        self.posts = []
        
    def add_posts(self, post):
        self.posts.append(post)
        print(f"Post : '{post}' added in Username : '{self.username}' Profile.")
        
    def display_post(self):
        print("\n Display Post")
        print(f" Username : {self.username}")
        for i in self.posts:
            print(f"{i}")

    def search_posts(self, keyword):
        matching_posts = [post for post in self.posts if keyword in post]
        if matching_posts:
            print(f"Posts by {self.username} containing '{keyword}':")
            for i, post in enumerate(matching_posts, start=1):
                print(f"{i}. {post}")
        else:
            print(f"No posts by {self.username} containing '{keyword}' found.")
        

U1 = Social_media("fmohd")
U2 = Social_media("Harry123")

U1.add_posts("If you run from the work, you might as well hidw from the money too")
U1.add_posts("mastering in full stck data science")
U1.add_posts("this is fsds pro batch")
U2.add_posts("A NICE CAR AND BIG HHOUSE are the old status symbol. The ultimate flex is freedom. Time freedom, location freedom and financial freedom")
U2.add_posts("To be successful at anything, don't have to be different. you simply have to be what most people aren't: consistent")
U1.display_post()
U1.search_posts("science")
U2.display_post()
```

```python
# Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates.
#             Implement methods to add tasks, mark tasks as completed, and display pending tasks.

class ToDo_List:              # create a class of ToDo list
    def __init__(self):       # create attribute tasks and due date
        self.tasks = {}
#        self.due_date = 0
        
    def add_task(self, tasks, due_date):  # implement method to add tasks
        self.tasks[tasks] = due_date
        print(f"Task : {tasks} :: Due Date : {due_date} added.")
        
    def mark_tasks(self, tasks):
        print(f"Given Task <{tasks}> completed.")   # implement method to mark task completed.
        self.tasks.pop(tasks)
        
    def display_pending_tasks(self):               # implement method to display pending tasks
        pending_task = [task for task in self.tasks]
        if pending_task:
            print(f"\n Pending Task: {self.tasks}")
        else:
            print("No tasks Available")

T1 = ToDo_List()      # create an object of class ToDo list

T1.add_task("array assignment", "20/10/23")
T1.add_task("pointer assignment", "30/10/23")
T1.add_task("linkedlist assignment", "20/11/23")

T1.display_pending_tasks()
T1.mark_tasks("pointer assignment")
print()
T1.display_pending_tasks()
```
