```python
# 1. What is a constructor in Python? Explain its purpose and usage.
'''
In Python, a constructor is a special method used to initialize and create objects of a class.
It is defined within a class and has the name __init__. The constructor is automatically called when an object of the class is created, 
and it is used to set up initial attributes and perform any necessary setup for the object.

In Python every class has a constructor; it's not required to define explicitly.
The purpose of the constructor is to construct an object and assign a value to the object's members

The purpose and usage of a constructor in Python are as follows:

1. Initialization: The primary purpose of a constructor is to initialize the attributes or properties of an object.
It allows you to set initial values for the object's state.

2. Customization: Constructors can accept parameters, enabling you to customize the initialization of each object.
This allows you to create objects with different initial states.

3. Automatically Invoked: Constructors are automatically called when you create an object of the class.
You don't need to call them explicitly; Python takes care of this for you.

4. Defining Default Values: You can define default values for attributes in the constructor, so if a value is not provided during object creation,
the object has sensible default values.
'''
class MyClass:                           # here we create a class
    def __init__(self, name, rollno):#we defined constructor and pass two attribute whenever we create an object we necessary to initialize attribute
        self.name = name
        self.rollno = rollno

# Creating an object of the class using the constructor
obj = MyClass("Farhan", 16)
print(obj.name)
print(obj.rollno)
```

```python
# 6. How can you call a constructor explicitly in Python? Give an example.
'''
In Python, you typically don't call a constructor explicitly. Constructors are automatically invoked when you create an object of a class
using the class name followed by parentheses. However, in some cases, you might want to call a constructor explicitly,especially 
if you're working with inheritance and want to call the constructor of a parent class. To do this, you can use the super() function.
'''

class ParentClass:
    def __init__(self, param1):
        self.param1 = param1

class ChildClass(ParentClass):
    def __init__(self, param1, param2):
        # Call the constructor of the parent class explicitly
        super().__init__(param1)
        self.param2 = param2

# Create an object of the ChildClass
child_obj = ChildClass("Value1", "Value2")

# Access the attributes of the child object
print(child_obj.param1)  # Output: "Value1" (inherited from ParentClass)
print(child_obj.param2)  # Output: "Value2" (defined in ChildClass)
```

```python
# 10. How can you have multiple constructors in a Python class? Explain with an example.
'''
Python does not support explicit multiple constructors, yet there are some ways using which multiple constructors can be achieved.
If multiple __init__ methods are written for the same class, then the latest one overwrites all the previous constructors and the 
reason for this can be, python stores all the function names in a class as key in a dictionary so, when a new function is defined 
with the same name, the key remains the same but the value gets overridden by the new function body.'''
class example:
    def __init__(self):
        print("One")
    def __init__(self):
        print("Two")
    def __init__(self):
        print("Three")
e = example()

'''
Multiple constructors are required when one has to perform different actions on the instantiation of a class. This is useful when
the class has to perform different actions on different parameters. The class constructors can be made to exhibit polymorphism in
three ways which are listed below.

1. Overloading constructors based on arguments.
2. Calling methods from __init__.
3. Using @classmethod decorator.'''

class sample:
	def __init__(self, *args):               # constructor overloading based on args
		if len(args) > 1:                    # if args are more than 1, sum of args
			self.ans = 0
			for i in args:
				self.ans += i
		elif isinstance(args[0], int):       # if arg is an integer, square the arg
			self.ans = args[0]*args[0]
            
		elif isinstance(args[0], str):       # if arg is string Print with hello
			self.ans = "Hello! "+args[0]+"."
s1 = sample(1, 2, 3, 4, 5)
print("Sum of list :", s1.ans)
s2 = sample(5)
print("Square of int :", s2.ans)
s3 = sample("Pwskills")
print("String :", s3.ans)
```

```python
# 16. How do you prevent a class from having multiple instances by using constructors in Python? Provide an example.
'''
In Python, you can prevent a class from having multiple instances by implementing a design pattern called the Singleton pattern. The Singleton 
pattern ensures that a class has only one instance, and it provides a global point of access to that instance. To achieve this in Python, you 
can use a combination of class attributes and a class method to control the creation of instances. Here's how you can implement the Singleton 
pattern to prevent a class from having multiple instances:
'''
class SingletonClass:
    _instance = None  # Class variable to store the single instance

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super(SingletonClass, cls).__new__(cls)
        return cls._instance

# Creating instances
instance1 = SingletonClass()
instance2 = SingletonClass()

# Checking if both instances are the same
print(instance1 is instance2)  # Output: True
```

```python
# 19. Explain the use of constructor chaining in Python. Provide a practical example.
'''
Constructor chaining, also known as constructor delegation, is a technique in object-oriented programming that allows one constructor of a class
to call another constructor of the same class.

Constructor chaining is useful when you want to invoke multiple constructors, one after another, by initializing only one instance. In Python,
constructor chaining is convenient when we are dealing with inheritance. When an instance of a child class is initialized, the constructors of all
the parent classes are first invoked and then, in the end, the constructor of the child class is invoked.
Using the super() method we can invoke the parent class constructor from a child class.
'''
class Vehicle:
    # Constructor of Vehicle
    def __init__(self, engine):
        print('Inside Vehicle Constructor')
        self.engine = engine

class Car(Vehicle):
    # Constructor of Car
    def __init__(self, engine, max_speed):
        super().__init__(engine)
        print('Inside Car Constructor')
        self.max_speed = max_speed

class Electric_Car(Car):
    # Constructor of Electric Car
    def __init__(self, engine, max_speed, km_range):
        super().__init__(engine, max_speed)
        print('Inside Electric Car Constructor')
        self.km_range = km_range

# Object of electric car
ev = Electric_Car('1500cc', 240, 750)
print(f'Engine={ev.engine}, Max Speed={ev.max_speed}, Km range={ev.km_range}')
```

```python
# 5. How can you access the methods and attributes of a parent class from a child class in Python? Give an example.
'''
Accessing Parent Class Functions
When a class inherits from another class it inherits the attributes and methods of another class. A class that inherits from another class
is known as child class and the class from which the child class inherits is known as Parent class. But have you ever wondered how to access
the parent’s class methods? This is really simple, you just have to call the constructor of parent class inside the constructor of child class
and then the object of a child class can access the methods and attributes of the parent class.
'''
'''
Here we create a super class called person and a child class called Employee, Employee class inherit the property of parent class which is Person.
we inherit the parent class attribute name and idnumber just have to call the constructor of parent class inside the constructor of child class and
then object of child class can access the the methods and attributes of the parent class.
'''

class Person:
    def __init__(self, name, IdNumber):
        self.name = name
        self.IdNumber = IdNumber
        
    def Display_Details(self):
        print(f"Name : {self.name} :: Id Number : {self.IdNumber}")
        
class Employee(Person):
    def __init__(self, name , IdNumber, Position):
        self.Position = Position
        
        Person.__init__(self, name, IdNumber)
        
    def Display(self):
        
        print(f"Name : {self.name} :: Id Number : {self.IdNumber} :: Position : {self.Position}")
        
E1 = Employee("Ayush", "D1234", "SDE")
E1.Display_Details()
E1.Display()

```

```python
# 12. Explain the use of abstract base classes (ABCs) in Python and how they relate to inheritance. Provide an example using the `abc` module.
'''
The main goal of the abstract base class is to provide a standardized way to test whether an object adheres to a given specification. It can 
also prevent any attempt to instantiate a subclass that doesn’t override a particular method in the superclass. And finally, using an abstract
class, a class can derive identity from another class without any object inheritance.

An abstract class can be considered as a blueprint for other classes.
Abstract classes are classes that contain one or more abstract methods. An abstract method is a method that has a declaration but not has a
implementation.
'''

from abc import ABC, abstractmethod

# Define an abstract base class (ABC)
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

# Create concrete subclasses that inherit from the abstract class
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

    def perimeter(self):
        return 2 * 3.14 * self.radius

class Square(Shape):
    def __init__(self, side_length):
        self.side_length = side_length

    def area(self):
        return self.side_length * self.side_length

    def perimeter(self):
        return 4 * self.side_length

# Using the classes
circle = Circle(5)
square = Square(4)

# Calling the methods
print("Circle Area:", circle.area())
print("Circle Perimeter:", circle.perimeter())

print("Square Area:", square.area())
print("Square Perimeter:", square.perimeter())
```

```python
# 2. Describe the key principles of encapsulation, including access control and data hiding.
'''
Encapsulation is one of the four fundamental Object-Oriented Programming (OOP) principles and involves bundling the data (attributes) and
methods (functions) that operate on the data into a single unit known as a class. The key principles associated with encapsulation are access
control and data hiding.

1. Access Control:-
 Access control refers to the restriction of access to certain components (attributes and methods) of a class. It helps in preventing unauthorized
 access and modification of an object's internal state.
 Access control is typically implemented through access modifiers, such as public, private, and protected.
 
 Public:- Members (attributes or methods) are accessible from outside the class.
 Private:- Members are only accessible within the class. Objects of the class cannot access private members directly.
 Protected:- Members are accessible within the class and its subclasses.
 
2. Data Hiding:-
 Data hiding refers to the practice of restricting direct access to certain attributes or implementation details of a class.
 It is achieved by making attributes private or providing controlled access through getter and setter methods.
 By hiding implementation details, the internal state of an object is protected, and changes to the internal representation can be made without
 affecting the code that uses the class.
'''
```

```python
# 1. Explain the concept of composition in Python and how it is used to build complex objects from simpler ones.
'''
In Python, composition is a design principle that allows complex objects to be built by combining simpler objects. It involves 
creating a new object by including other objects as its components or parts, rather than through inheritance. Composition enables 
developers to create complex and flexible structures while promoting code reusability, modularity, and maintainability.

Combining Objects:- 
Composition involves creating a new object and including other objects as its attributes. These included objects are typically 
instances of other classes, and they represent the various components or parts of the composite object.

Has-A Relationship:- 
Composition establishes a "has-a" relationship between the composite object and its components. This means that the composite 
object contains or is composed of instances of other classes, rather than inheriting behavior from them.
'''
class Engine:
    def start(self):
        print("Engine started.")

class Wheels:
    def rotate(self):
        print("Wheels rotating.")

class Transmission:
    def shift_gear(self):
        print("Transmission shifting gear.")

class Car:
    def __init__(self):
        self.engine = Engine()
        self.wheels = Wheels()
        self.transmission = Transmission()

    def drive(self):
        self.engine.start()
        self.transmission.shift_gear()
        self.wheels.rotate()

# Usage:
my_car = Car()
my_car.drive()
```

```python
# 6. Create a Python class hierarchy for a music player system, using composition to represent playlists and songs.

class Song:
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist = artist
        self.duration = duration

    def play(self):
        print(f"Playing '{self.title}' by {self.artist} ({self.duration} seconds)")


class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []

    def add_song(self, song):
        self.songs.append(song)

    def play_all(self):
        print(f"Playing playlist: {self.name}")
        for song in self.songs:
            song.play()


class MusicPlayer:
    def __init__(self):
        self.playlists = []

    def add_playlist(self, playlist):
        self.playlists.append(playlist)

    def play_playlist(self, playlist_name):
        for playlist in self.playlists:
            if playlist.name == playlist_name:
                playlist.play_all()
                return
        print(f"Playlist '{playlist_name}' not found")


# Example usage:
song1 = Song("Shape of You", "Ed Sheeran", 240)
song2 = Song("Bohemian Rhapsody", "Queen", 354)
song3 = Song("Hotel California", "Eagles", 390)

playlist1 = Playlist("Pop Hits")
playlist1.add_song(song1)
playlist1.add_song(song2)

playlist2 = Playlist("Classic Rock")
playlist2.add_song(song2)
playlist2.add_song(song3)

music_player = MusicPlayer()
music_player.add_playlist(playlist1)
music_player.add_playlist(playlist2)

music_player.play_playlist("Pop Hits")
music_player.play_playlist("Classic Rock")
music_player.play_playlist("Nonexistent Playlist")
```

```python
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id

    def get_info(self):
        return f"Student: {self.name} (ID: {self.student_id})"


class Instructor:
    def __init__(self, name, instructor_id):
        self.name = name
        self.instructor_id = instructor_id

    def get_info(self):
        return f"Instructor: {self.name} (ID: {self.instructor_id})"


class CourseMaterial:
    def __init__(self, textbook, slides):
        self.textbook = textbook
        self.slides = slides

    def get_info(self):
        return f"Textbook: {self.textbook}, Slides: {self.slides}"


class UniversityCourse:
    def __init__(self, course_name, instructor, students, course_material):
        self.course_name = course_name
        self.instructor = instructor
        self.students = students
        self.course_material = course_material

    def get_course_info(self):
        course_info = f"Course: {self.course_name}\n"
        course_info += f"Instructor: {self.instructor.get_info()}\n"
        course_info += "Students:\n"
        for student in self.students:
            course_info += f"- {student.get_info()}\n"
        course_info += f"Course Material: {self.course_material.get_info()}"
        return course_info


# Example usage:
student1 = Student("Rajat", "S1")
student2 = Student("Avinash", "S2")
instructor = Instructor("Ashish Vishnoi", "M1")
course_material = CourseMaterial("Python Programming", "Week 1 slides")
students = [student1, student2]

course = UniversityCourse("Introduction to Python", instructor, students, course_material)
print(course.get_course_info())
```

```python
# 13. Discuss the challenges and drawbacks of composition, such as increased complexity and potential for tight coupling between objects.
'''
While composition offers numerous benefits, such as modularity, code reusability, and flexibility, it also presents challenges 
and drawbacks that developers need to be mindful of:

Increased Complexity:-
Using composition can lead to increased complexity in the codebase, especially as the number of composed objects and their 
interactions grow. Managing the relationships between objects and coordinating their behavior can become more challenging, 
making the code harder to understand and maintain.

Potential for Tight Coupling:- 
Composition can result in tight coupling between objects, especially if the containing object directly accesses and manipulates 
the internal details of its composed objects. Tight coupling can make the code less flexible and more resistant to changes, 
as modifications to one object may require changes to other dependent objects.

Complex Object Creation:- 
Objects composed of multiple components may require complex initialization logic, leading to more complicated object creation 
processes. Managing the creation and configuration of composed objects can become cumbersome, especially if there are multiple 
levels of composition involved.

Potential Performance Overhead:- 
Depending on the implementation, using composition may introduce some performance overhead, as it involves additional object 
creation and method delegation. While this overhead is often negligible, it may become significant in performance-critical applications.

To mitigate these challenges and drawbacks, developers should carefully design their composition hierarchies, strive for loose 
coupling between objects, encapsulate behavior within objects, and adhere to best practices in software design. Additionally, 
thorough testing and documentation can help ensure the correctness and maintainability of composed objects in the codebase.
'''
```

```python
# 14. Create a Python class hierarchy for a restaurant system, using composition to represent menus, dishes, and ingredients.

class Ingredient:
    def __init__(self, name, quantity, unit):
        self.name = name
        self.quantity = quantity
        self.unit = unit

    def display(self):
        return f"{self.quantity} {self.unit} of {self.name}"
    
class Dish:
    def __init__(self, name, description, ingredients):
        self.name = name
        self.description = description
        self.ingredients = ingredients
        
    def display(self):
        dish_info = f"{self.name}: {self.description}\nIngredients:"
        for ingredient in self.ingredients:
            dish_info += f"\n- {ingredient.display()}"
        return dish_info
    
class Menu:
    def __init__(self, name, dishes):
        self.name = name
        self.dishes = dishes

    def display(self):
        menu_info = f"{self.name} Menu\nDishes:"
        for dish in self.dishes:
            menu_info += f"\n{dish.display()}"
        return menu_info
    
class Restaurant:
    def __init__(self, name, menus):
        self.name = name
        self.menus = menus
    def display_menu(self):
        restaurant_info = f"Welcome to {self.name}!\nAvailable Menus:"
        for menu in self.menus:
            restaurant_info += f"\n{menu.display()}"
        return restaurant_info
        

ingredient1 = Ingredient("Tomato", 2, "pieces")
ingredient2 = Ingredient("Cheese", 150, "grams")
ingredient3 = Ingredient("Bread", 2, "slices")

dish1 = Dish("Caprese Salad", "Fresh tomato and mozzarella salad", [ingredient1, ingredient2])
dish2 = Dish("Grilled Cheese Sandwich", "Classic comfort food", [ingredient2, ingredient3])

menu1 = Menu("Lunch Specials", [dish1, dish2])

restaurant = Restaurant("Globe Resturants", [menu1])
print(restaurant.display_menu())
```

```python
# 16. Create a Python class for a computer game character, using composition to represent attributes like weapons, armor, and inventory.
class Weapon:
    def __init__(self, name, damage):
        self.name = name
        self.damage = damage

    def display(self):
        return f"Weapon: {self.name}, Damage: {self.damage}"

class Armor:
    def __init__(self, name, defense):
        self.name = name
        self.defense = defense

    def display(self):
        return f"Armor: {self.name}, Defense: {self.defense}"

class Inventory:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)

    def display(self):
        if not self.items:
            return "Inventory is empty."
        inventory_info = "Inventory:\n"
        for item in self.items:
            inventory_info += f"- {item.display()}\n"
        return inventory_info

class GameCharacter:
    def __init__(self, name, weapon, armor, inventory):
        self.name = name
        self.weapon = weapon
        self.armor = armor
        self.inventory = inventory

    def display_character_info(self):
        character_info = f"{self.name}\n"
        character_info += f"{self.weapon.display()}\n"
        character_info += f"{self.armor.display()}\n"
        character_info += f"{self.inventory.display()}"
        return character_info

# Example usage:
weapon = Weapon("Sword of Bahubali", 30)
armor = Armor("Dragonhide Armor", 20)
inventory = Inventory()
inventory.add_item(Weapon("Bow of Precision", 25))
inventory.add_item(Armor("Steel Plate Armor", 15))

character = GameCharacter("Bahubali", weapon, armor, inventory)
print(character.display_character_info())
```

```python
# 18. Create a Python class for a house, using composition to represent rooms, furniture, and appliances.

class Furniture:
    def __init__(self, name, color):
        self.name = name
        self.color = color

    def description(self):
        return f"{self.color} {self.name}"

class Appliance:
    def __init__(self, name, brand):
        self.name = name
        self.brand = brand

    def description(self):
        return f"{self.brand} {self.name}"

class Room:
    def __init__(self, name, area):
        self.name = name
        self.area = area
        self.furniture = []
        self.appliances = []

    def add_furniture(self, furniture):
        self.furniture.append(furniture)

    def add_appliance(self, appliance):
        self.appliances.append(appliance)

    def display(self):
        room_info = f"{self.name} ({self.area} sq.ft)\n"
        room_info += "Furniture:\n"
        for furniture in self.furniture:
            room_info += f"- {furniture.description()}\n"
        room_info += "Appliances:\n"
        for appliance in self.appliances:
            room_info += f"- {appliance.description()}\n"
        return room_info

class House:
    def __init__(self):
        self.rooms = []

    def add_room(self, room):
        self.rooms.append(room)

    def display(self):
        house_info = "House Information:\n"
        for room in self.rooms:
            house_info += f"{room.display()}\n"
        return house_info

# Example usage:
living_room = Room("Living Room", 300)
living_room.add_furniture(Furniture("Sofa", "Brown"))
living_room.add_furniture(Furniture("Coffee Table", "Black"))
living_room.add_appliance(Appliance("TV", "Samsung"))

kitchen = Room("Kitchen", 200)
kitchen.add_furniture(Furniture("Dining Table", "Wooden"))
kitchen.add_appliance(Appliance("Refrigerator", "LG"))
kitchen.add_appliance(Appliance("Oven", "Whirlpool"))

house = House()
house.add_room(living_room)
house.add_room(kitchen)

print(house.display())
```
